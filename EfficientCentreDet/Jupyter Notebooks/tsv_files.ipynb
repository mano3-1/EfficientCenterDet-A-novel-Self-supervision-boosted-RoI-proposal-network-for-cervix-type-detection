```python
def random_viz(tsv,i):
    k = random.randint(0,tsv.shape[0])
    string = tsv.iloc[k].to_list()[0][7:]
    string = string.split(' ')
    img_dir = '/home/b170007ec/Datasets/Cervix Cancer/train/train/train/Type_'+str(i)+'/'+string[0]
    bbox = string[2:6]
    for i in range(4):
        bbox[i] = int(bbox[i])
    print(bbox)
    img = cv2.imread(img_dir)
    print(img.shape)
    cv2.rectangle(img,(bbox[0],bbox[1]),(bbox[0]+bbox[2],bbox[1]+bbox[3]),(0,0,0),10)
    plt.imshow(img)
```

```python
random_viz(tsv3,3)
```
